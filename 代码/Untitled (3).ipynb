```python
from matplotlib import font_manager

fonts = sorted(set(f.name for f in font_manager.fontManager.ttflist))
for name in fonts:
    if any(k in name.lower() for k in ['noto', 'cjk', 'wenquan', 'yahei', 'hei', 'song', 'kai', 'fang', 'arial unicode']):
        print(name)
```
